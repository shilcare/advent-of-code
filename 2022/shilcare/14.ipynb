```python
def sandFallDown(grid, sandStartPos):
    numSands = 0
    numSandsComeToRest = 0
    n, m = len(grid), len(grid[0])
    fallingForever = False
    while not fallingForever:
        i, j = sandStartPos
        comeToRest = False
        numSands += 1
        while not comeToRest and not fallingForever:
            while i + 1 < n:
                if grid[i + 1][j] == ".":  # fall down
                    i += 1
                elif j - 1 < 0:  # falling forever to the left
                    fallingForever = True
                    break
                elif grid[i + 1][j - 1] == ".":  # falling diagonally down and left
                    i += 1
                    j -= 1
                elif j + 1 >= m:  # falling forever to the right
                    fallingForever = True
                    break
                elif grid[i + 1][j + 1] == ".":  # falling diagonally down and right
                    i += 1
                    j += 1
                else:
                    comeToRest = True
                    break
            if i + 1 >= n:
                fallingForever = True

        if comeToRest:
            numSandsComeToRest += 1
            grid[i][j] = "o"
        # if numSands < 25:
        #     for row in grid:
        #         print(''.join(row))
        #     print("Sands: ", numSands)
        # else:
        #     break
    return numSandsComeToRest


with open("14.input", "r") as f:
    minX = minY = float("inf")
    maxX = maxY = float("-inf")
    paths = []
    for line in f:
        path = line.strip().split(" -> ")
        paths.append([])
        for coordinate in path:
            x, y = map(int, coordinate.split(","))
            paths[-1].append((x, y))
            minX = min(minX, x)
            maxX = max(maxX, x)
            minY = min(minY, y)
            maxY = max(maxY, y)
    # print(minX, maxX, minY, maxY)
    sandStartPos = (500, 0)
    # grid[0][0] is the top left corner at (minX, 0)
    grid = [["." for _ in range(maxX - minX + 1)] for _ in range(maxY + 1)]
    n, m = len(grid), len(grid[0])
    grid[0][500 - minX] = "+"  # sand start position
    for path in paths:
        for i in range(len(path) - 1):
            x1, y1 = path[i]
            x2, y2 = path[i + 1]
            if x1 == x2:
                yStart, yEnd = sorted([y1, y2])
                for y in range(yStart, yEnd + 1):
                    grid[y][x1 - minX] = "#"
            else:
                xStart, xEnd = sorted([x1, x2])
                for x in range(xStart, xEnd + 1):
                    grid[y1][x - minX] = "#"

    numSands = sandFallDown(grid, (0, 500 - minX))
    print(numSands)
```

```python
def sandFallDownWithFloor(grid, sandStartPos):
    numSands = 0
    numSandsComeToRest = 0
    n, m = len(grid), len(grid[0])
    sandBlocked = False
    while not sandBlocked:
        i, j = sandStartPos
        comeToRest = False
        numSands += 1
        while not comeToRest:
            while i + 1 < n:
                if grid[i + 1][j] == ".":  # fall down
                    i += 1
                elif j - 1 < 0:  # falling forever to the left, should not happen
                    raise Exception("Falling forever to the left")
                elif grid[i + 1][j - 1] == ".":  # falling diagonally down and left
                    i += 1
                    j -= 1
                elif j + 1 >= m:  # falling forever to the right, should not happen
                    raise Exception("Falling forever to the right")
                elif grid[i + 1][j + 1] == ".":  # falling diagonally down and right
                    i += 1
                    j += 1
                else:
                    if (i, j) == sandStartPos:
                        sandBlocked = True
                    comeToRest = True
                    break
            if i + 1 >= n:
                raise Exception("Falling forever without falling down to the floor")

        if comeToRest:
            numSandsComeToRest += 1
            grid[i][j] = "o"
    return numSandsComeToRest


with open("14.input", "r") as f:
    minX = minY = float("inf")
    maxX = maxY = float("-inf")
    paths = []
    for line in f:
        path = line.strip().split(" -> ")
        paths.append([])
        for coordinate in path:
            x, y = map(int, coordinate.split(","))
            paths[-1].append((x, y))
            minX = min(minX, x)
            maxX = max(maxX, x)
            minY = min(minY, y)
            maxY = max(maxY, y)
    # print(minX, maxX, minY, maxY)
    sandStartPos = (500, 0)
    minX -= maxY
    maxX += maxY
    floorYDeta = 2
    grid = [["." for _ in range(maxX - minX + 1)] for _ in range(maxY + 1 + floorYDeta)]
    n, m = len(grid), len(grid[0])
    grid[0][500 - minX] = "+"  # sand start position
    for path in paths:
        for i in range(len(path) - 1):
            x1, y1 = path[i]
            x2, y2 = path[i + 1]
            if x1 == x2:
                yStart, yEnd = sorted([y1, y2])
                for y in range(yStart, yEnd + 1):
                    grid[y][x1 - minX] = "#"
            else:
                xStart, xEnd = sorted([x1, x2])
                for x in range(xStart, xEnd + 1):
                    grid[y1][x - minX] = "#"
    # rock path on the floor
    for j in range(m):
        grid[-1][j] = "#"

    numSands = sandFallDownWithFloor(grid, (0, 500 - minX))
    print(numSands)
```
