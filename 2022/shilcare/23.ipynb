```python
def printElves(elves: set[(int, int)], round: int):
    boundaries = (
        min(x for x, y in elves),
        max(x for x, y in elves),
        min(y for x, y in elves),
        max(y for x, y in elves),
    )
    for y in range(boundaries[3] - boundaries[2] + 1):
        for x in range(boundaries[1] - boundaries[0] + 1):
            if (x + boundaries[0], y + boundaries[2]) in elves:
                print("#", end="")
            else:
                print(".", end="")
        print()
    print(f"==== Round {round} ====", end="\n\n")
```

```python
from collections import defaultdict
from itertools import count

elves = set()
with open("23.input", "r") as f:
    data = f.read().splitlines()
    for y, line in enumerate(data):
        for x, c in enumerate(line):
            if c == "#":
                elves.add((x, y))

# initial order of directions: north, south, west, east
directions = [
    [(-1, -1), (0, -1), (1, -1)],  # NW, N, NE
    [(-1, 1), (0, 1), (1, 1)],  # SW, S, SE
    [(-1, -1), (-1, 0), (-1, 1)],  # NW, W, SW
    [(1, -1), (1, 0), (1, 1)],  # NE, E, SE
]
startDirection = 0
totalEleves = len(elves)

for round in count():
    newElves = defaultdict(list)
    numElvesMoved = 0
    for x, y in elves:
        neighbors = set((x + dx, y + dy) for d in directions for dx, dy in d)
        if all(pos not in elves for pos in neighbors):
            newElves[(x, y)].append((x, y))
            continue
        proposed = False
        for i in range(len(directions)):
            direction = directions[(startDirection + i) % len(directions)]
            if all((x + dx, y + dy) not in elves for dx, dy in direction):
                proposed = True
                numElvesMoved += 1
                proposedPos = (x + direction[1][0], y + direction[1][1])
                newElves[proposedPos].append((x, y))
                break
        if not proposed:
            newElves[(x, y)].append((x, y))

    elves = set()
    for pos, collisions in newElves.items():
        if len(collisions) == 1:
            elves.add(pos)
        else:
            numElvesMoved -= len(collisions)
            for x, y in collisions:
                elves.add((x, y))
    # printElves(elves, round + 1)
    startDirection = (startDirection + 1) % 4

    if round == 9:
        boundaries = (
            min(x for x, y in elves),
            max(x for x, y in elves),
            min(y for x, y in elves),
            max(y for x, y in elves),
        )
        area = (boundaries[1] - boundaries[0] + 1) * (boundaries[3] - boundaries[2] + 1)
        print(area - len(elves))

    print(f"round {round + 1}: {numElvesMoved} elves moved")

    if numElvesMoved == 0:
        print("no Elf moves at round", round + 1)
        break
```
